```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('P:\coding\OfficeDataProject.csv')
df.head()
```

```python
df.describe()
```

```python
sns.pairplot(df[['age', 'salary']])
```

```python
import sklearn.cluster as cluster
```

```python
kmeans = cluster.KMeans(n_clusters=5 ,init="k-means++")
kmeans = kmeans.fit(df[['age','salary']])
```

```python
kmeans.cluster_centers_
```

```python
df['Clusters'] = kmeans.labels_
df.head()
```

```python
df['Clusters'].value_counts()
```

```python
sns.scatterplot(x="age", y="salary",hue = 'Clusters',  data=df)
```
